```python
def frequency_distribution(df):
    numeric_columns = df.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        print(f"Frequency distribution for {column}:")
        print(df[column].value_counts().sort_index())
        print("\n")
frequency_distribution(df)
```

```python
import matplotlib.pyplot as plt
def plot_histograms(df):
    numeric_columns = df.select_dtypes(include=['number']).columns
    for column in numeric_columns:
        plt.figure()
        df[column].hist(bins=100)
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {column}')
        plt.show()

# Calling the function
plot_histograms(df)
```

```python
import seaborn as sns

numeric_columns = df.select_dtypes(include=['number']).columns

# Plot scatter plots
for column in numeric_columns:
    if column != 'Response':
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=df[column], y=df['Response'])
        plt.title(f'Scatter plot of {column} vs Response')
        plt.xlabel(column)
        plt.ylabel('Response')
        plt.show()
```

```python
from scipy.stats import chi2_contingency

# Function to perform chi-square test for all categorical columns against the response variable
def chi_square_test(df, Response):
    categorical_columns = df.select_dtypes(include=['object', 'category']).columns
    for column in categorical_columns:
        if column != Response:
            contingency_table = pd.crosstab(df[Response], df[column])
            chi2, p, dof, expected = chi2_contingency(contingency_table)
            print(f"Chi-square test for {column} vs {Response}:")
            print(f"Chi2: {chi2}, p-value: {p}\n")

# Calling the function
chi_square_test(df, 'Response')
```

```python
import numpy as np
from datetime import datetime
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'],errors='coerce')
df['Age'] = np.where(df['Age'] > 90, 90, df['Age'])
df['Customer_age']=datetime.now().year-df['Dt_Customer'].dt.year
```

```python
# Separate categorical columns
categorical_cols = df.select_dtypes(include=['object','category']).columns

# Perform one-hot encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols)

y=df_encoded['Response']
X=df_encoded.drop(['Response','Dt_Customer','Year_Birth','Id'],axis=1)
```

```python
!pip install scikit-optimize
import xgboost as xgb
from sklearn.model_selection import train_test_split
from skopt import BayesSearchCV


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=44)
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')

# Define the parameter search space
param_space = {
    'n_estimators': (50, 200),
    'max_depth': (3, 10),
    'learning_rate': (0.01, 0.2, 'log-uniform'),
    'sub_sample': (0.8, 1.0)
}

# Perform Bayesian optimization
bayes_search = BayesSearchCV(estimator=xgb_model, search_spaces=param_space, n_iter=100, cv=5, scoring='roc_auc')
bayes_search.fit(X_train, y_train)

# Print the best parameters and the best score
print("Best parameters found: ", bayes_search.best_params_)
print("Best accuracy score: ", bayes_search.best_score_)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=44)
xgb_model = xgb.XGBClassifier(use_label_encoder=False,learning_rate=0.05255647132234826,
                        n_estimators=181,max_depth=3,sub_sample=.8,      eval_metric='mlogloss')


xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)
from sklearn.metrics import f1_score
# Calculate F1 score
y_pred = xgb_model.predict(X_test)
f1 = f1_score(y_test, y_pred, average='micro')
auc = roc_auc_score(y_test, y_pred, average='micro')
print(f"test F1 Score: {f1}")

print(f"test auc Score: {auc}")
y_pred = xgb_model.predict(X_train)
f1 = f1_score(y_train, y_pred, average='micro')
auc = roc_auc_score(y_train, y_pred, average='micro')
print(f"train F1 Score: {f1}")

print(f"train auc Score: {auc}")
```

```python
from sklearn.model_selection import KFold, cross_val_score
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
cv_results = cross_val_score(xgb_model, X, y, cv=kf, scoring='roc_auc')

# Print cross-validation results
print(f'Cross-validation accuracy scores: {cv_results}')
print(f'Mean cross-validation roc_auc: {cv_results.mean()}')

cv_results = cross_val_score(xgb_model, X, y, cv=kf, scoring='f1_micro')

# Print cross-validation results
print(f'Cross-validation accuracy scores: {cv_results}')
print(f'Mean cross-validation f1: {cv_results.mean()}')
```

```python
import shap
# Create a SHAP explainer
explainer = shap.Explainer(xgb_model)

# Calculate SHAP values for the test set
shap_values = explainer(X_train)

# Plot the SHAP values for the first instance in the test set
shap.plots.waterfall(shap_values[0])

# Plot the summary plot for all instances in the test set
shap.summary_plot(shap_values, X_train)
shap.summary_plot(shap_values, X_train, plot_type="bar")
```
